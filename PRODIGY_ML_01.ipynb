```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data=pd.read_csv("housing.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
#splitting data into training and testing data 

from sklearn.model_selection import train_test_split
x=data.drop(['median_house_value'],axis=1)  #data exceptt the target (predicted value)
y=data['median_house_value']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
train_data=x_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(15,8))
#The histograms will be displayed as a set of bars, where each bar represents a range of values (bin), and the height of the bar represents the frequency or count of values falling within that bin.
```

## Correlation
correlation coefficient of -1 indicates a perfect negative correlation. In the heatmap, it would be represented by a dark color, often in a shade of red or blue depending on the chosen color map.
A correlation coefficient of 1 indicates a perfect positive correlation. This, too, would be represented by a dark color, but on the opposite side of the color spectrum compared to the negative correlation.
A correlation coefficient of 0 indicates no correlation. In the heatmap, it would be represented by a lighter color, often closer to the middle of the color spectrum.

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
```

```python
train_data['total_rooms']=np.log(train_data['total_rooms']+1)
train_data['total_bedrooms']=np.log(train_data['total_bedrooms']+1)
train_data['population']=np.log(train_data['population']+1)
train_data['households']=np.log(train_data['households']+1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data=train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=train_data, hue='median_house_value',palette='coolwarm')
```

```python
#engineering some features that are interesting
train_data['bedrooms_ratio']=train_data['total_bedrooms']/train_data['total_rooms']
train_data['household']=train_data['total_rooms']/train_data['households']
```

```python
#train simple linear regression
from sklearn.linear_model import LinearRegression 
x_train,y_train=train_data.drop(['median_house_value'],axis=1),train_data['median_house_value']
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
#scale the data 
test_data=x_test.join(y_test)
test_data['total_rooms']=np.log(test_data['total_rooms']+1)
test_data['total_bedrooms']=np.log(test_data['total_bedrooms']+1)
test_data['population']=np.log(test_data['population']+1)
test_data['households']=np.log(test_data['households']+1)

test_data=test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)

test_data['bedrooms_ratio']=test_data['total_bedrooms']/test_data['total_rooms']
test_data['household']=test_data['total_rooms']/test_data['households']
```

```python
x_test,y_test=test_data.drop(['median_house_value'],axis=1),test_data['median_house_value']
```
